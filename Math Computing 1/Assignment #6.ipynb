### Exercise 38

#### Using list comprehension, create an expression that sums up all the factors of a number that you input. (For Example, for input=6, output=12 (1+2+3+6=12)

```python
#input number you wish to sum up
#convert input to integer
#create a list of factors of the input integer
#create for loop to find factors of input
#sum up the factors and output

print('Sum up the Factors of a Number')

x=int(input('Enter a number whose factors you wish to sum: '))

factors=[a for a in range(1,x+1) if x%a==0]

sum1=0
for a in factors:
    sum1+=a
print(sum1)
```

#### Write a function named 'transform' that takes as arguments list1, list2, r1, and r2, that removes items from list1 in the slice r1:r2, appends them onto list2 in reverse order, and returns the resulting list.

```python
#prompt for two lists with commas
#separate the elements in the list
#create a function 'transform'

list1=input("Enter your first list with comma separation: ")
list2=input("Enter your second list with comma separation: ")

list1=list1.split(',')
list2=list2.split(',')

r1=int(input("Enter first index of slice: "))
r2=int(input("Enter second index of slice: "))

def transform(list1, list2, r1, r2):
    list1=list1.remove[r1:r2]
    list1=list1.reverse()
    list3=list2.extend(list1)
    return list3


print(transform(list1,list2,r1,r2))
```
