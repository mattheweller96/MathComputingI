#### Design a class called Sentence that has a constructor that takes a string representing a sentence as input. The class should have methods:
* get_first_word
* get_all_words
* replace(index, new_word): change a word at a index to new_word
#### Ex: sentence "I'm going back" and set_word_at_index(2, "home"), becomes "I'm going home".

```python
#1.split and strip the sentence so it is an index of words
#2.def(first)index at 0 of words is first word, return
#2.def(allword)all words is just the sentence itself
#3.def(replace)for the given index and new word, split the sentence 
#4.the sentence at the given index is the new word, return the sentence

class Sentence():
    
    def get_first_word(self,sentence):
        
        self.sentence = sentence
        words = sentence.strip().split()
        first = words[0]
        
        return first
    
    def get_all_words(self):
        
        return self.sentence
    
    def replace(self,index,new_word):
        
        self.index = index
        self.new_word = new_word
        
        sentence = self.sentence.split()
        sentence[index] = new_word
        
        return (sentence)
    
sentence = Sentence()


print(sentence.get_first_word("Brian Eno makes great music"))

print(sentence.get_all_words())

print(sentence.replace(1,"May"))
```
